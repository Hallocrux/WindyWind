```python
a = df[df['code'].str.startswith('W')].groupby('code') #.aggregate({'value1': 'mean', 'value2': 'mean', 'value3': 'mean'})
```

```python
import matplotlib.pyplot as plt

value_cols = ["value1", "value2", "value3"]

for group_code in a:
    g = df[df["code"] == group_code]
    if g.empty:
        continue

    cols_to_plot = [c for c in value_cols if c in g.columns and g[c].notna().any()]
    if not cols_to_plot:
        continue

    use_time = "time" in g.columns and g["time"].nunique() > 1
    x = g["time"] if use_time else g.index

    plt.figure(figsize=(12, 4))
    for col in cols_to_plot:
        plt.plot(x, g[col], label=col, linewidth=0.8)

    plt.title(f"{group_code} 各通道曲线")
    plt.xlabel("time" if use_time else "index")
    plt.ylabel("value")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```
